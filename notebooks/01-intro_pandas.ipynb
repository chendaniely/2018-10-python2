```python
import seaborn as sns
```

```python
# use the load_dataset function from seaborn
# and pass in the 'tips' value to load the tips dataset
tips = sns.load_dataset('tips')
```

```python
# look at the head of tips
tips.head(n=1)
```

```python
# subset total bill, tip, sex (1 dataset w/ 3 columns)
tips[['total_bill', 'tip', 'sex']]
```

```python
# total bill, tip, sex == male (1 dataset w/columns, but filtered)
tips.loc[tips['sex'] == 'Male', ['total_bill', 'tip', 'sex']]
```

```python
tips.loc[tips.sex == 'Male', ['total_bill', 'tip', 'sex']]
```

```python
# average tip
tips['tip'].mean()
```

```python
# average tip by sex
tips.groupby('sex')['tip'].mean()
```
