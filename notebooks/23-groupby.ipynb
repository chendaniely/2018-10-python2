```python
import pandas as pd
```

```python
df = pd.read_csv('../data/gapminder.tsv', sep='\t')
```

```python
df.head()
```

```python
df.groupby('continent').lifeExp.describe()
```

```python
import numpy as np
```

```python
df.groupby('continent').lifeExp.aggregate(np.mean)
```

```python
df.groupby('continent')['lifeExp'].agg(np.mean)
```

```python
def my_mean(values):
    n = len(values)
    sum = 0
    for v in values:
        sum += v # sum = sum + v
    return sum / n
```

```python
df.groupby('continent')['lifeExp'].agg(my_mean)
```

```python
df.groupby('continent')['lifeExp'].agg([np.count_nonzero, np.mean, np.std])
```

```python
df.groupby('continent')['lifeExp'].agg({
    'lifeExp': 'mean',
    'pop': 'median',
    'gdpPercap': 'median'
})
```

```python
def my_zscore(x):
    return (x - x.mean()) / x.std()
```

```python
from scipy.stats import zscore
```

```python
df.groupby('year')['lifeExp'].transform(zscore)
```
